```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
data
```

```python
x=data.drop('class',axis=1)
y=data['class']
```

```python
label_encoder=LabelEncoder()
y_encoder=label_encoder.fit_transform(y)
y_encoder
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y_encoder,test_size=0.3,random_state=42)
```

```python
classifier=RandomForestClassifier(n_estimators=100,random_state=4)
```

```python
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
accuracy=accuracy_score(y_test,y_pred)
print(f'Accuracy:{accuracy*100:.2f}%')
```

```python
y_pred_decoded=label_encoder.inverse_transform(y_pred)
print(f'Predicted_label:{y_pred_decoded}')
```

```python
new_data=pd.DataFrame({
    'pelvic_incidence':[60.5],
    'pelvic_tilt numeric':[21.3],
    'lumbar_lordosis_angle':[40.2],
    'sacral_slope':[45.3],
    'pelvic_radius':[110.2],
    'degree_spondylolisthesis':[3.5]
})
predictions=classifier.predict(new_data)
predicted_labels=label_encoder.inverse_transform(predictions)
print(f'Predicted Labels for new data:{predicted_labels}')
```
