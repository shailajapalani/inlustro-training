```python
import seaborn as sns
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Gender",data=df)
plt.title("Number of Gender",fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x="Gender",y="Survived",data=df,palette="dark")
plt.title("Number of Survivors based on Age",fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Age"],bins=30,kde=True)
plt.title("Age",fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df[["Age","Fare"]],palette="coolwarm")
plt.title("Boxplot for outliers")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.violinplot(x="Pclass",y="Age",data=df)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="PassengerId",y="Pclass",data=df)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.pie(x="Pclass",data=df)
plt.show()
```
